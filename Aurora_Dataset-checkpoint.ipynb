#### Visualisations necessary for Imputations

```python
# Assuming df3 is our DataFrame
# Replace the column names with the actual column names from your DataFrame
columns_to_plot = ['gse_x_ace', 'gse_y_ace', 'gse_z_ace', 'gse_x_dscovr', 'gse_y_dscovr', 'gse_z_dscovr']

# Select the specified columns
subset_df = df3[columns_to_plot]

# Create individual scatter plots for each pair of coordinates between ACE and DSCOVR
for i in range(0, len(columns_to_plot)//2):
    plt.figure(figsize=(2.5, 2))
    ace_column = columns_to_plot[i]
    dscovr_column = columns_to_plot[i + len(columns_to_plot)//2]
    
    plt.scatter(subset_df[ace_column], subset_df[dscovr_column])
    plt.title(f'Scatter Plot of {ace_column} vs {dscovr_column}')
    plt.xlabel(f'{ace_column}')
    plt.ylabel(f'{dscovr_column}')
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd

# Create a DataFrame with 'day' column ranging from 0 to 2435
days_repeated = sorted(list(range(2436)) * 24)
test = pd.DataFrame({'day': days_repeated})

# Create a 'time' column repeating 0 to 23 hours for each day
hours_repeated = [hour % 24 for hour in range(len(test))]
test['time'] = hours_repeated

# Create a timestamp column by combining 'day' and 'time'
test['timestamp'] = pd.to_datetime(test['day'], unit='D') + pd.to_timedelta(test['time'], unit='H')

test
```
